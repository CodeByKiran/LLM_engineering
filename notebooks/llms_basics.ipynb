```python
import google.generativeai as genai
import os
```

```python
# Configuration of API key
genai.configure(api_key=os.getenv("gemini_api_key"))


# Model Selection
model = genai.GenerativeModel("gemini-3-flash-preview")

#Sample Sentence to check the no.of tokens 
sample = "The TurboBlend Pro 5000 stopped working after two weeks."

# Count tokens
token_info = model.count_tokens(sample)

```

```python
print(f"Text: {sample}")
print(f"Token count: {token_info.total_tokens}")
```

Find the Number of Token in a Sample of 10 reviews

```python
import pandas as pd

df = pd.read_csv("../data/raw/homenest_reviews.csv")
df.columns
```

```python
reviews = df['review_text'].head(10)

token_counts = []
for i , reviews in enumerate(reviews, start =1 ):
    token_info = model.count_tokens(reviews)


    print(f"\nReview {i}")
    print(f"Text : {reviews}")
    print(f"Token Count : {token_info.total_tokens}")
    token_counts.append(token_info.total_tokens)

print(f"\n\nMin : {min(token_counts)}")  
print(f'Max : {max(token_counts)}') 
print(f'Avg :{sum(token_counts) /len(token_counts)}')
```

Temperature Experiments

```python

import sys
sys.path.append("../src")

from llm_client import call_llm

review = """I purchased the TurboBlend Pro 5000 with high expectations because it was advertised as a powerful and durable blender suitable for everyday kitchen use. Initially, it worked quite well and handled basic blending tasks without any issues. However, after just two weeks of regular use, the blender suddenly stopped working completely. The motor appears to have died, and the appliance will not turn on anymore. This was extremely disappointing, especially considering the product is supposed to be a premium model. I expected much better reliability and longevity from a product in this category. Overall, my experience has been frustrating, and I would hesitate to recommend the TurboBlend Pro 5000 to others unless the manufacturer improves its build quality and durability.
"""

system = "You are a product review analyst. Summarize the given review into a one line sentence."

results = []
for temp in [0.0, 0.7, 0.8, 1.0]:
    for run in range(3):
        out = call_llm(system, review, temperature=temp,provider = "mistral")
        results.append((temp, run+1, out))

for r in results:
    print(f"Temp={r[0]} | Run={r[1]} | Summary={r[2]}")
```

```python
import pandas as pd
import time
import sys
sys.path.append("../src")

from llm_client import call_llm


reviews = [
    "The TurboBlend Pro 5000 was a massive disappointment. I bought it specifically for heavy-duty green smoothies, and for the first ten days, it handled frozen kale and ice with ease. However, on day fourteen, the unit emitted a loud pop and a smell of burning electronics; the motor died completely right in the middle of a cycle. For a 'Pro' model at this price point, a two-week lifespan is unacceptable and points to serious quality control issues.",
    "I have mixed feelings about this coffee maker. On the positive side, it brews incredibly quickly and the flavor extraction is top-tier—the coffee tastes great and stays hot. However, the physical construction is a letdown. The plastic build feels cheap and thin, the water reservoir lid creaks every time I open it, and the overall housing feels like it could crack if it's bumped too hard on the counter.",
    "I was genuinely worried when the delivery arrived because the outer packaging was heavily damaged and crushed on one side. It looked like it had been dropped or stepped on during transit. Thankfully, the internal foam bracing was thick enough that the product itself works fine and arrived without a single scratch. The manufacturer gets an A+ for protective internal packaging, even if the courier was careless.",
    "After using several high-end smartphones over the years, I can say the battery life on this model is truly impressive. I am a power user who spends hours on GPS navigation and video calls, yet the device lasts all day even with heavy use. I usually head to bed with 30% battery remaining, which is a huge relief compared to my previous phone that required a mid-afternoon charge every single day.",
    "While I actually like the product itself, the company's customer support was very slow and unhelpful when I reached out. I waited on hold for forty minutes only to be told I had to submit a manual ticket, and the email responses I received were robotic and failed to address my specific request for a replacement part. It’s frustrating when a good product is overshadowed by a complete lack of helpful post-purchase service."
]

system = "You are product review analyst. Summrize the given Reviews into one line review Summary"
```

```python
results = []

for review in reviews:
    
    claude_out =call_llm(system,review,provider='claude')
    mistral_out =call_llm(system,review,provider='mistral')
    
    results.append({
        "Review": review,
        "Claude Response":claude_out,
        "Claude Length": len(claude_out.split()),
        "Mistral Response": mistral_out,
        "Mistral Length": len(mistral_out.split())
    })

df = pd.DataFrame(results)

df
```
